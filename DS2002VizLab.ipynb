```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/hulu.csv')


# first few rows
print(df.head())

# info about the dataframe
print(df.info())

# summary statistics
print(df.describe())

# creating visualizations
# bar chart of categorical data
top_10_genres = df['genres'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_genres.index, top_10_genres.values, color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.title('Top 10 Genres on Hulu')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['releaseYear'],df['imdbNumVotes'], alpha=0.5)
plt.xlabel('Release Year')
plt.ylabel('Number of IMDB Votes')
plt.title('Release Year vs. Number of IMDB Votes')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/sample_data/hulu.csv')

# first few rows of the dataframe
print(df.head())

# getting information about the dataframe
print(df.info())

# Summary statistics
print(df.describe())


# Visualizations using Seaborn
# 1. Countplot for genres
plt.figure(figsize=(12, 6))
sns.countplot(y='genres', data=df, order=df['genres'].value_counts().iloc[:10].index, palette='viridis')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.title('Top 10 Genres on Hulu')
plt.show()

# 5. Heatmap of correlations between numerical features
numerical_features = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
